# STUDYING JSONFILE

```python
print("NUMBER OF DISTRICTKEYS IN JSON FILE--",len(jdata.keys()))
```

```python
modifiedjsonnames = []   #LIST CONTAINING WITHOUT "_district" and "_" BEWTWEEN DISTRICT NAMES
modified_keys={}          # CREATING A MAPPING USING DICTIONARY   (original name mapped to required name)

for i in jdata.keys():
    
    #MAKING NECESSARY CHANGES
    t = i.replace('_district','')     
    t=t.replace('_',' ')
    t=t.replace('–',' ')
    t=t.replace('-',' ')
    t=t.rsplit('/')[0]

    modified_keys[i]=t                           
    modifiedjsonnames.append(t)
```

```python
modifiedjsonnames[:5]   #CONSISTS OF CORRECT SPELLINGS
```

```python
for i in range(len(ls_neighbourjson)):             #LOOPING ON WRONG SPELLING
    for j in range(len(modifiedjsonnames)):        #FINDING CORRECT SPELLING IN "MODIFIEDJSONNAMES"
        if ls_neighbourjson[i] == modifiedjsonnames[j]:  # if we find
            for key,val in modified_keys.items():
                if val == modifiedjsonnames[j]:
                    modified_keys[key] = ls_districtwise[i]
            modifiedjsonnames[j]=ls_districtwise[i]     # replacing with correct spelling from excel
```

```python
vacc_dist_list = vacc_data['District'].to_list()
vacc_dist_key = vacc_data['District_Key'].to_list()

newname = []     #IT CONSISTS OF FULLNAME i.e."STATEID_DISTRICTNAME"
```

```python
count=0
for i in range(len(modifiedjsonnames)):
    for j in range(len(vacc_dist_list)):
        if modifiedjsonnames[i] == vacc_dist_list[j]:
#             print(vacc_dist_key[j])
            newname.append(vacc_dist_key[j])
```

```python
newname[:5]
```

```python
#AS ABOVE, DISTRICT CONSISTS OF CAPITAL LETTERS WE LOWER THEM
st_dist = []
for i in newname:
    t = i.rsplit('_')
#     name = t[1]
    lower = t[1].lower()
    st_dist.append([t[0],lower])
```

APPNDING STATEID AND DISTRICTNAME INTO OUR DICTIONARY MAPPING "MODIFIED_KEYS"

```python
for key,val in modified_keys.items():
    for i in st_dist:
        if i[1] == val:
            name = i[0]+'_'+i[1]
            modified_keys[key] = name
```

```python
# new_dict = {}           #OUR FINAL DICTIONARY
# for key,val in data_json.items():
#     v = modified_keys[key]
#     new_list = []
#     for i in val:
#         z=modified_keys[i]
#         new_list.append(z)
```

```python
new_dict = {}
for key,val in data_json.items():
    v = modified_keys[key]
    
#     print(val)
    new_list = []
    for i in val:
        z=modified_keys[i]
        new_list.append(z)
    new_dict[v] = new_list
```

```python
len(new_dict.keys())
```

```python
#DISTRICTS THAT ARE NOT PRESENT IN VACCINE DATA
districts_to_be_deleted=['konkan division', 'niwari', 'noklak','mumbai suburban','ri bhoi']
```

```python
#DELETING THE ABOVE DISTRICTS FROM THE  KEYS DICTIONARY

for i in districts_to_be_deleted:
    del new_dict[i]
```

```python
#DELETING LIST-VALUES IN THE DICTIONARY
for key , value in new_dict.items():
    value1=[]
    for i in value: 
        if i in districts_to_be_deleted:
            pass
        else:
            value1.append(i)
    new_dict[key] = value1
        
```

## FINAL JSON OUTPUT FILE

```python
from collections import OrderedDict
```

```python
#SORTING ON DICT-KEYS
dict1 = OrderedDict(sorted(new_dict.items()))
```

```python
finaldict={}
#SORTING ON DICT-VALUE LISTS
for key , value in dict1.items():
    finaldict[key] = sorted(value)
```

```python
with open('neighbor-districts-modified.json','w') as q:
    json.dump(finaldict,q,indent=5)
```
